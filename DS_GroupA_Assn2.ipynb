```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'Student_id' : range(1,31),
    'Study_Hours': np.random.normal(5,2,30),
    'Attendance' : np.random.normal(75,20,30),
    'Previous_Grade' : np.random.normal(70,10,30),
    'Final_Grade' : np.random.normal(70,10,30)
})

data.loc[18,'Study_Hours'] = np.nan
data.loc[6,'Attendance'] = np.nan
data.loc[10,'Previous_Grade'] = np.nan
data.loc[25,'Final_Grade'] = np.nan
```

```python
data.fillna(data.mean(),inplace=True)
```

```python
plt.xlabel('Study Hours')
plt.ylabel('Final Grade')
plt.scatter(data['Study_Hours'],data['Final_Grade'])
plt.show()
```

```python
Q1 = data['Attendance'].quantile(0.25)
Q3 = data['Attendance'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data['Attendance'] < lower_bound) | (data['Attendance'] > upper_bound)]
print("Outliers:\n")
outliers
```

```python
columns = ['Study_Hours', 'Attendance', 'Previous_Grade',
       'Final_Grade']
def outlier_treatment(data,column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Corrected line to cap outliers
  data[column] = np.where(data[column] < lower_bound, lower_bound,
                          np.where(data[column] > upper_bound, upper_bound, data[column]))
  # The commented-out lines below were for printing outliers, which is not the capping operation.
  # outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
  print(f"{column}:\n")
  print(f"Lower Bound: {lower_bound}")
  print(f"Upper Bound: {upper_bound}")
  # print(outliers)
  return data
for column in columns:
  data = outlier_treatment(data,column)
```
