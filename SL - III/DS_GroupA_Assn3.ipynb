```python
import pandas as pd
#importing the pandas library
```

```python
#reading the 'titanic.csv' file and storing it in a dataframe
data = pd.read_csv('/content/titanic.csv')
data #print the dataframe
```

```python
data.columns
#all the columns of the dataframe
```

```python
#grouped by 'pclass'
#Performed (mean, median, min, max, standard deviation) on columns[age, fare]
grouped_stats = data.groupby("pclass")[["age", "fare"]].agg(
    ["mean", "median", "min", "max", "std"]
)
grouped_stats
```

```python
#creating a list grouping it by pclass = [1,2,3]
#storing the mean of p[i] at the ith index of the mean_fare_list
mean_fare_list = data.groupby("pclass")["fare"].mean().tolist()
print(mean_fare_list)
```

```python
#read and stores the Iris.csv file in a dataframe
data2 = pd.read_csv('/content/Iris.csv')
data2
```

```python
def calculate_stats(category):
  print(f"---------{category}----------")
  print(data2.groupby('Species')[category].describe())  #grouped by species and performed statistical insights
  print("\n")                                           # on each category of the the 3 species of the flowers

columns = data2.columns #columns of the dataset
for column in columns:
  calculate_stats(column) #passed every column as an argument in the function
```
