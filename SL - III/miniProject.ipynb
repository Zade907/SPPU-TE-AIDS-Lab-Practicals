```python
import pandas as pd
import re
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

nltk.download('vader_lexicon')

df = pd.read_csv(r"C:\Users\athni\Downloads\tweets\data_science.csv", low_memory=False)

# Use correct column
df = df[['tweet']].dropna().head(100)

# Clean text
def clean(text):
    text = re.sub(r"http\S+|@\w+|#", "", str(text))
    text = re.sub(r"[^A-Za-z ]", "", text)
    return text.lower()

df['clean'] = df['tweet'].apply(clean)

# Sentiment
sia = SentimentIntensityAnalyzer()
df['score'] = df['clean'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Classification
df['sentiment'] = df['score'].apply(lambda x: "Positive" if x >= 0 else "Negative")

# Output
print(df[['tweet','sentiment']].head())
print(df['sentiment'].value_counts())
```
