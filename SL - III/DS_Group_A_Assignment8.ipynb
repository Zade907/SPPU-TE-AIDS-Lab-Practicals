```python
import seaborn as sns
import matplotlib.pyplot as plt

# load inbuilt dataset
titanic = sns.load_dataset('titanic')

# view first rows
print(titanic.head())

```

```python
# ---------------------------
# Pattern 1: Survival by Gender
# ---------------------------
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival Count by Gender")
plt.show()

```

```python
# ---------------------------
# Pattern 2: Survival by Passenger Class
# ---------------------------
sns.countplot(x='class', hue='survived', data=titanic)
plt.title("Survival by Passenger Class")
plt.show()

```

```python
# ---------------------------
# Pattern 3: Age Distribution
# ---------------------------
sns.histplot(titanic['age'].dropna(), bins=30)
plt.title("Age Distribution")
plt.show()

```

```python
# ---------------------------
# Histogram for Fare Distribution
# ---------------------------
sns.histplot(titanic['fare'], bins=30)
plt.title("Fare Distribution")
plt.xlabel("Ticket Fare")
plt.ylabel("Passenger Count")
plt.show()
```
