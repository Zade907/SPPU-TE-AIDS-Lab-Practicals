```python
# Sample log data
logs = [
    "2026-04-21 INFO Login successful",
    "2026-04-21 ERROR Database failed",
    "2026-04-21 INFO File uploaded",
    "2026-04-21 WARNING Disk space low",
    "2026-04-21 ERROR Timeout error"
]

# MAP FUNCTION
def mapper(logs):
    mapped = []
    for line in logs:
        parts = line.split()
        log_level = parts[1]   # INFO / ERROR / WARNING
        mapped.append((log_level, 1))
    return mapped

# SHUFFLE + SORT
def shuffle(mapped_data):
    shuffled = {}
    for key, value in mapped_data:
        if key not in shuffled:
            shuffled[key] = []
        shuffled[key].append(value)
    return shuffled

# REDUCE FUNCTION
def reducer(shuffled_data):
    reduced = {}
    for key in shuffled_data:
        reduced[key] = sum(shuffled_data[key])
    return reduced

# Run MapReduce
mapped = mapper(logs)
shuffled = shuffle(mapped)
result = reducer(shuffled)

print("Log Level Count:")
print(result)
```
